```python
import pandas as pd
melb_path='data/melb_data.csv' #Đường dẫn đến file data
melb_data=pd.read_csv(melb_path) # Đọc file data
melb_data.columns # In ra các cột để kiểm tra
```

```python
#Dữ liệu trong tệp data có một số trường bị thiếu " Missing value " . Nếu giữ nguyên tệp data này sẽ ảnh hưởng đến kết quả training của model 
#Vì vậy, chúng ta cần loại bỏ các missing value này
melb_data=melb_data.dropna(axis=0) #dropna(axis=0): Xóa các hàng (rows) chứa giá trị NaN (giá trị bị thiếu, không có dữ liệu). Nếu đặt axis=1 thì loại bỏ các cột
```

```python
#Mục tiêu của model là dự đoán giá nhà. Vì vậy chúng ta sẽ tạo biến y và lưu giá trị giá nhà trong tệp data vào biến y này
y=melb_data.Price
```

```python
#Các cột trong tệp data sẽ được gọi là các Features. Chúng ta thấy trong tệp data có rất nhiều Features
#Vì vậy , chúng ta sẽ chọn lại các Features nào có khả năng ảnh hưởng đến giá nhà
melb_Features=['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
#Theo quy ước thì dữ liệu này được gọi là X
X=melb_data[melb_Features]

```

```python
#Review lại data trước khi train
X.describe()
```

```python
#Sử dụng phương thức head để xem 1 vài hàng trên cùng
X.head(5)
```

```python
#Bước tiếp theo sẽ là xây dựng model . Gôm các giai đoạn sau
#1.Define: Xác định loại model
#2.Fit : sử dụng data đã xử lý để train cho model 
#3.Predict: sử dụng model đã train để dự đoán kết quả
#4.Evaluate : xác định độ chính xác của model
#Trong ví dụ này , sẽ sử dụng thư viện  scikit-learn và sử dụng loại model decision tree 
from sklearn.tree import DecisionTreeRegressor
melb_model=DecisionTreeRegressor(random_state=1) # Định nghĩa loại model và xác định tham số random_state =1 để đảm bảo kết quả giống nhau mỗi lần chạy
#fit model 
melb_model.fit(X,y) 
```

```python
#Trong thực tế thì sẽ sử dụng model để dự đoán những giá nhà mới
#Trong trường họp này sẽ sử dụng lại những hàng đầu tiên của tệp data để xem mô hình hoạt động đúng chưa
print("Dự đoán giá nhà của 5 hàng đầu trong tệp data")
print(X.head())
#Dự đoán giá nhà
print("Giá nhà là : ")
print(melb_model.predict(X.head()))
```
